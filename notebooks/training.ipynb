# Training Visualization

This notebook plots the training and validation loss curves from the 10-epoch training run.

```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('../training_history.json') as f:
    history = json.load(f)

train_loss = history['train_loss']
val_loss = history['val_loss']
epochs = list(range(1, len(train_loss) + 1))

print(f'Epochs: {len(epochs)}')
print(f'Final train loss: {train_loss[-1]:.4f}')
print(f'Final val loss:   {val_loss[-1]:.4f}')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Loss curves
ax1.plot(epochs, train_loss, 'o-', label='Train Loss', color='#2196F3')
ax1.plot(epochs, val_loss, 's-', label='Validation Loss', color='#FF5722')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Cross-Entropy Loss')
ax1.set_title('Training and Validation Loss')
ax1.legend()
ax1.set_xticks(epochs)
ax1.grid(True, alpha=0.3)

# Perplexity curves
train_ppl = [np.exp(l) for l in train_loss]
val_ppl = [np.exp(l) for l in val_loss]

ax2.plot(epochs, train_ppl, 'o-', label='Train PPL', color='#2196F3')
ax2.plot(epochs, val_ppl, 's-', label='Validation PPL', color='#FF5722')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Perplexity')
ax2.set_title('Training and Validation Perplexity')
ax2.legend()
ax2.set_xticks(epochs)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('training_curves.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Train-validation gap over epochs
gap = [v - t for t, v in zip(train_loss, val_loss)]

plt.figure(figsize=(7, 4))
plt.bar(epochs, gap, color='#9C27B0', alpha=0.7)
plt.xlabel('Epoch')
plt.ylabel('Val Loss - Train Loss')
plt.title('Generalization Gap per Epoch')
plt.xticks(epochs)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print(f'Gap epoch 1:  {gap[0]:.4f}')
print(f'Gap epoch 10: {gap[-1]:.4f}')
```
